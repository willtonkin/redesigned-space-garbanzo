```python
import folium
from folium.plugins import TimestampedGeoJson
import pandas as pd
import io
import ipywidgets as widgets
from IPython.display import display

# Define the test data as a string variable
data_str = '''title,lat,lon,date,year
Central Park,40.7829,-73.9654,1900-01-01,1900
Empire State Building,40.7484,-73.9857,1931-05-01,1931
Statue of Liberty,40.6892,-74.0445,1886-10-28,1886
One World Trade Center,40.7127,-74.0134,2013-11-03,2013
Chrysler Building,40.7517,-73.9753,1930-05-27,1930
Times Square,40.7589,-73.9851,1904-12-31,1904'''

# Read the data from the string variable
df = pd.read_csv(io.StringIO(data_str))

# Create a map object
map_obj = folium.Map(location=[40.7128, -74.0060], zoom_start=12)

# Define the function that generates the GeoJSON data for a given year
def generate_data(year):
    data = {
        'type': 'FeatureCollection',
        'features': []
    }
    for index, row in df.iterrows():
        if row['year'] == year:
            feature = {
                'type': 'Feature',
                'geometry': {
                    'type': 'Point',
                    'coordinates': [row['lon'], row['lat']]
                },
                'properties': {
                    'title': row['title'],
                    'date': row['date']
                }
            }
            data['features'].append(feature)
    return data

# Define the slider control using ipywidgets
slider = widgets.IntRangeSlider(
    value=[1900, 2022],
    min=1900,
    max=2022,
    step=1,
    description='Year range:',
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# Define the callback function for the slider control
def slider_callback(change):
    year_range = [int(year) for year in change['new']]
    data = generate_data(year_range[0])
    geojson.add_data(data)

# Add the callback function to the slider control
slider.observe(slider_callback, names='value')

# Display the slider control
display(slider)

# Define the TimestampedGeoJson layer
geojson = TimestampedGeoJson(
    data={},
    period='P1Y',
    duration='P1M',
    auto_play=False,
    loop=False,
    max_speed=1,
    loop_button=True,
    date_options='YYYY-MM-DD',
    time_slider_drag_update=True
)

# Add the TimestampedGeoJson layer to the map object
map_obj.add_child(geojson)

# Display the map object
display(map_obj)
```
